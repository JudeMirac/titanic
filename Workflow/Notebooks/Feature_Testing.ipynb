```python
# Load libraries
import pandas as pd

# upload dataset with engineered features
df1 = pd.read_csv('uncut_feat')
```

```python
# before traing let's view which features are important in predicitng the target variable 

# Download ppscore 
!pip install ppscore

# import ppscore
import ppscore as pps

# convert target variable to object type 
df1['Survived'] = df1['Survived'].astype(str)
```

```python
# calculate PPS for the engineered feature
interaction1 = pps.score(df1, 'interaction1', 'Survived')
# Calculate PPS for individual variables 
score_sibsp = pps.score(df1, 'SibSp', 'Survived') 
score_parch = pps.score(df1, 'Parch', 'Survived')
# print results 
score_sibsp, score_parch, interaction1

# if the score is higher than the interaction feature, then it must be removed due to iniformational dilution
# (In otherwords, combining the variables togther may have weakend the predicitive power of the other leading to nosie creation)
```
